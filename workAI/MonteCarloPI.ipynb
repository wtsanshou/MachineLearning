```python
from pylab import randn
import matplotlib.pyplot as plt
import numpy as np
import math

axes = plt.axes()
axes.set_aspect('equal', 'datalim')


axes.set_xlim([-1, 1])
axes.set_ylim([-1, 1])

axes.grid()

size = 50

X = np.random.uniform(-1.0,1.0,size)
Y = np.random.uniform(-1.0,1.0,size)

num = 0
for i in range(size):
    if math.sqrt(X[i]*X[i] + Y[i]*Y[i])<=1.0:
        num += 1

print("ratio = Area of Circle / Area of Square")     
print("ratio = PI * r^2       / (2*r)^2") 
print("ratio = PI / 4") 
print("PI = ratio * 4") 
print()
print("Monte Carlo Simulation result:")
print("Number of inside dots 'Area of Circle': ", num)
print ("PI is ", 4, "*",num, "/",size, " = ", 4*num/size)
plt.scatter(X,Y)
circle1=plt.Circle((0,0),1.0,color='g', alpha=0.5)
plt.gcf().gca().add_artist(circle1)
plt.show()
```

```python

import matplotlib
%matplotlib inline
matplotlib.style.use('ggplot')

np.random.seed(14543)

def showDistribution(mean, sd, size):
    data = np.random.normal(mean, sd, size)

    count, bins, ignored = plt.hist(data, 30, range=(0,10))
    plt.plot(bins, 1/(sd * np.sqrt(2 * np.pi)) *
            np.exp( - (bins - mean)**2 / (2 * sd**2) ),
            linewidth=2)
    plt.show()
print('Player A')
showDistribution(5, 2, 1000)

print('Player B')
showDistribution(7, 2, 1000)
```
